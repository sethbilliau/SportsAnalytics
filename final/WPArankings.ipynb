# Get WPA Rankings 

```python
url = 'https://wpapool.com/ranking/' 
tables = pd.read_html(url)
```

```python


# print(playerlst)
WPAmen = tables[0]
WPAmen.Player[np.where(WPAmen.Player== "Chris Robinson-Reinhold")[0][0]] = "Chris Robinson"
playerlst = list(WPAmen.Player)
playerlst

WPArankings = [" ".join(n.split(", ")[::-1]) for n in playerlst]
WPAmen["Player"] 
WPAmenrankingsdict = dict(zip(WPArankings, WPAmen["Total Points"]))

# rankingsdict
```

```python
mens_rankings = pd.DataFrame.from_dict(WPAmenrankingsdict,orient='index', columns = ["rankings"])
```

```python
# Women
WPAwomen = tables[1]
playerlst = list(WPAwomen.Player)
playerlst
WPArankings = [" ".join(n.split(", ")[::-1]) for n in playerlst]
WPAwomen["Player"] 
WPAwomenrankingsdict = dict(zip(WPArankings, WPAwomen["Total Points"]))
```

```python
womens_rankings = pd.DataFrame.from_dict(WPAwomenrankingsdict, orient='index', columns = ["rankings"])
womens_rankings
```

# Full Rankings

```python
fullrankings = pd.concat([womens_rankings, mens_rankings]).sort_values(by="rankings", ascending = False)
```

```python
fullrankings.reset_index(inplace=True)
fullrankings = fullrankings.rename(columns = {'index':'player'})
fullrankings
```

```python
fullrankings['player'] = deduplicate_names(fullrankings['player'])
fullrankings
```

```python
fullrankings.to_csv("data/WPArankings.csv")
```
